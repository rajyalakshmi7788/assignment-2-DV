3.Low level data understanding:

```python
 count_value = data["location"].nunique()
print("Count value of Location = ", count_value)
```

```python
continent_frequency = data['continent'].value_counts()
max_continent = continent_frequency.idxmax()
print("Maximum value of frequency in",max_continent)
```

```python
max_of_total_case = data["total_cases"].max()
print("Maximum of total cases: ",max_of_total_case)
mean_of_total_case = data["total_cases"].mean()
print("Mean of total cases: ",mean_of_total_case)
```

```python
quartile_25 = data['total_deaths'].quantile(0.25)
quartile_50 = data['total_deaths'].quantile(0.5)
quartile_75 = data['total_deaths'].quantile(0.75)
print("25% quartile value total_death :",quartile_25)
print("50% quartile value total_death :",quartile_50)
print("75% quartile value total_death :",quartile_75)
```

```python
df = data.loc[data["human_development_index"].idxmax(),"continent"]
print(df," has maximum human development index")
```

```python
df = data.loc[data["gdp_per_capita"].idxmin(),"continent"]
print(df," has minimum gdp_per_capita")
```

```python
min_value = data.groupby("continent")["gdp_per_capita"].agg("min")
min_val = min_value.idxmin()
print(min_val," has minimum GDP per capita")
```

```python
data = data.drop_duplicates()
data.head()
```
